```python
import pandas as pd
import math 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
from collections import Counter
%matplotlib inline
```

```python
fifa = pd.read_csv('fifa21_train.csv')
fifa.shape
```

```python
fifa_data = fifa.copy()
```

```python
def clean_data(function_name):
    import numpy as np
    import pandas as pd

    #Changing columns' names
    cols = []
    for column in function_name.columns:
        cols.append(column.lower())
    function_name.columns = cols
    
    function_name = function_name.drop(['id', 'loan date end'], axis =1)
    function_name['club'] = function_name['club'].replace(np.nan,'Unknown')
    function_name['joined'] = function_name['joined'].replace(np.nan,'Jan 1/2019')
    function_name = function_name.dropna()
    
    function_name.reset_index(inplace=True, drop=False)
    function_name = function_name.drop(['index'], axis =1)
    
    for cols1 in function_name.loc[:,'ls':'gk']:
        function_name[cols1] = list(map(lambda x: x[:2], function_name[cols1]))
        function_name[cols1] = function_name[cols1].str.replace('+', '.')
        function_name[cols1] = pd.to_numeric(function_name[cols1])
        
    function_name['gk'] = function_name['gk'].astype(int)
    
    for cols1 in function_name[['w/f','sm','ir']]:
        function_name[cols1] = list(map(lambda x: x[:1], function_name[cols1]))
        function_name[cols1] = pd.to_numeric(function_name[cols1])
    
    for i in range(len(function_name)):
        if function_name['hits'][i].count('K') >= 1:
            function_name['hits'][i] = function_name['hits'][i].replace('K', '')
            function_name['hits'][i] = pd.to_numeric(function_name['hits'][i])
            function_name['hits'][i] = function_name['hits'][i]*1000
        else:
            function_name['hits'][i] = pd.to_numeric(function_name['hits'][i])
    function_name['hits'] = pd.to_numeric(function_name['hits'])      
    
    cleaned_data = function_name
    return cleaned_data
```

```python
cleaned = clean_data(fifa_data)
cleaned
```
